# Ridge Regression
Ridge regression is a type of regularized regression model. This means it is a variation of the standard linear regression model that includes a regularized term in the cost function. The purpose of this is to prevent Overfitting. Ridge Regression adds an L2 regularization term to the linear equation. Thatâ€™s why it is also known as L2 Regularization or L2 Norm.

T**he main aim of Ridge Regression is to reduce Overfitting.**

When you fit a Linear Regression model, here is what happens. You have a set of features (often called X and represented as a matrix) and you want to find a set of coefficients (often called Î² and represented as a vector) by which you multiply the values in X to predict your target (often called y and represented as a vector)

The problem is that, in some cases, a Linear Regression will overfit specific datasets. What do you do in this case? Use Ridge and Lasso regression.

How do these models work?
Lasso and Ridge are both Linear Regression models but with a penalty (also called a regularization). They add a penalty to how big your beta vector can get, each in a different way.

```python
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

# Applying regRegression

```python
from sklearn.linear_model import Ridge
#if we inrease the value of alpha(lamda) it decrease the value of m
rr=Ridge(alpha=100)
rr.fit(X,y)
print(rr.coef_)
print(rr.intercept_)
```

```python
rr1 = Ridge(alpha=200)
rr1.fit(X,y)
print(rr1.coef_)
print(rr1.intercept_)
```

```python
plt.plot(X,y,'b.')
plt.plot(X,lr.predict(X),color='red',label='alpha=0')
plt.plot(X,rr.predict(X),color='green',label='alpha=10')
plt.plot(X,rr1.predict(X),color='orange',label='alpha=100')
plt.legend()
```

## Make your own class for linear regression

```python
def linear_regression(X,y,alpha=1):
  x_mean=X.mean()
  y_mean=y.mean()
  num=0
  den=0
  for i in range(X.shape[0]):
    num=num + (y[i] -y_mean) *(X[i] - x_mean)
    den=den + (X[i] -x_mean) *(X[i] - x_mean)
  m=num/(den+alpha )
  b=y_mean - m*x_mean
  return m,b
```
